### POURCENTAGE DE DONN2ES NULLES

##### tri des données pour insertion_df

```python
#reperer le pourcentage de valeurs nulles par variable : 

#petite fonction :

def percentage_null_df (df):
    return(df.isnull().sum() * 100 / len(df)).round(2) 
    # df.isnull() renvoie un dataframe dans lequel chaque valeur nulle est remplacée dans la tableau par "True" (et les non-nulles par "False")
    # or, True est aussi '1' (False est '0') : on peut réaliser des opérations avec.
    # par défaut, sum() somme selon l'axe 0, i.e. la verticale. l'objet renvoyé est une série (Pandas.Series)
    # cette fonction renvoie donc une série.
    # len(df) est le nombre de lignes (longueur selon la verticale, i.e. l'axe par défaut)
    # on obtient donc bien le pourcentage de lignes vides.

```

```python
nan_per_df = percentage_null_df(insertion_df)
display(nan_per_df)
```

```python
#nous n'allons pas parcourir toutes les variables, essayons avec missigno 

import missingno as miss
```

##### visualisation des données manquantes de insertion_df

```python
#colonnes nulles et visionage avec missigno

null_cols = nan_per_df.loc[lambda x : x>0].index.tolist()

nan_vals_df = insertion_df[null_cols]
miss.matrix(nan_vals_df, figsize=(10, 8))
plt.show()

#display(null_cols)
```
